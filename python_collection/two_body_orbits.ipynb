```python
# """
# Orbit Calculations and Plot with Absolute RA and DEC
# """

def stars_orbits(star_keys, utils, octo_utils):
    """
    Plot projected orbits for multiple stars on the sky (RA/Dec, mas).

    Parameters:
    -----------
    star_keys : List of star identifiers (e.g., ["A", "B", "C"]) to plot.
    utils : Utility functions and parameters for orbital calculations.
    octo_utils : Contains star data and cluster parameters.
    """

    # Create figure and axis for plotting
    fig, ax = plt.subplots(figsize=(8, 8))

    # Cluster center coordinates (converted from deg to mas)
    ra_cm_mas  = octo_utils.ra_cm_mas   # [mas]
    dec_cm_mas = octo_utils.dec_cm_mas  # [mas]

    # Distance to cluster and conversion factor [km], [mas/km]
    dist_km = octo_utils.distance_km.value
    km_to_mas = (1 / dist_km) * 206_265_000  # mas per km conversion

    # Lists to store all velocity LineCollections and projected velocities
    lc_vel_list = []  
    star_data = []    

    # Initialize global velocity min and max for color normalization
    v_proj_min = np.inf
    v_proj_max = -np.inf

    # First pass: calculate orbits and velocities for each star to get global velocity range
    for star_key in star_keys:
        star = octo_utils.stars[star_key]

        # Absolute Right Ascension and Declination (mas)
        ra  = star.ra    # [mas]
        dec = star.dec   # [mas]

        # Proper motions [mas/yr]
        pm_ra = star.pm_ra*u.mas/u.yr
        pm_dec = star.pm_dec*u.mas/u.yr

        # 2D  velocity [km/s]
        v2D = star.v2D *u.km/u.s

        #  === Orbital parameters ===

        # Distance to Omega Centauri center (km)
        distance_kpc = 5.43 * u.kpc
        distance_km= distance_kpc.to(u.km)
        # v_sample is a sample orbital speed using stars A and C from Omega Centauri
        v_sample = 100*u.km/u.s
        # a_sample uses the average angular distance from stars A and C in Omega Centauri to derive a sample semi-major axis in kilometers
        # approximately 0.5" at a cluster distance of 5.43kpc
        angular_distance_rad = (0.5*u.arcsec).to(u.rad).value
        a_sample = angular_distance_rad*distance_km
        # alternative a_sample values using AU as imput
        # a_sample = (12*u.au).to(u.km)
        
        # For KeplerEllipse
        e      = 0.75                                  # Eccentricity [unitless]
        a      = a_sample.value          # Semi-major axis [km]
        per    = utils.circular_period(semi_major=a_sample, speed=v_sample).value    # Period (s)
        tau    = 0                                     # Time of periapsis passage [days]
        Omega  = 50                                     # Longitude of ascending node [deg]
        w      = -80                                     # Argument of periapsis [deg]
        i      = 50                                     # Inclination [deg]

        # Time points for orbit sampling [yr]
        t = np.linspace(tau, tau + per, utils.n_int)

        # Create Keplerian orbit and get 3D positions [km]
        ke = pyasl.KeplerEllipse(a=a, per=per, e=e, Omega=Omega, i=i, w=w)
        pos = ke.xyzPos(t)
        ascn, descn = ke.xyzNodes_LOSZ()

        # Convert 3D orbit positions to RA/Dec offsets from cluster center [mas]
        ra_offset  = -pos[:, 1] * km_to_mas  # RA increases to the left on sky
        dec_offset =  pos[:, 0] * km_to_mas  # DEC increases upward

        # Absolute RA/Dec orbit positions [mas]
        ra_abs  = ra_cm_mas  + ra_offset
        dec_abs = dec_cm_mas + dec_offset

        # Line segments for orbit coloring by velocity
        points = np.array([ra_abs, dec_abs]).T.reshape(-1, 1, 2)
        segments = np.concatenate([points[:-1], points[1:]], axis=1)

        # Compute projected velocity magnitude [km/s]
        vel = ke.xyzVel(t)
        v_proj = np.sqrt(vel[:, 0]**2 + vel[:, 1]**2)

        # Update global velocity min and max for normalization
        v_proj_min = min(v_proj_min, v_proj.min())
        v_proj_max = max(v_proj_max, v_proj.max())

        # Store data for plotting after normalization is set
        star_data.append({
            "star_key": star_key,
            "ra": ra,
            "dec": dec,
            "pm_ra": pm_ra,
            "pm_dec": pm_dec,
            "v2D": v2D,
            "segments": segments,
            "v_proj": v_proj,
            "pos": pos,
            "ra_abs": ra_abs,
            "dec_abs": dec_abs,
        })

    # Create a common Normalize instance for projected velocity coloring
    norm_vel = Normalize(vmin=v_proj_min, vmax=v_proj_max)
    cmap = plt.get_cmap('plasma')

    # Plot the cluster center marker once
    ax.plot(ra_cm_mas, dec_cm_mas, 'k+', markersize=10, label='Central Mass')

    # Second pass: plot orbits, proper motion, and markers for each star using common normalization
    for data in star_data:
        star_key = data["star_key"]
        ra = data["ra"]
        dec = data["dec"]
        pm_ra = data["pm_ra"]
        pm_dec = data["pm_dec"]
        v2D = data["v2D"]
        segments = data["segments"]
        v_proj = data["v_proj"]
        pos = data["pos"]

        # Line collection colored by projected velocity [km/s]
        lc_vel = LineCollection(segments, cmap=cmap, norm=norm_vel)
        lc_vel.set_array(v_proj)
        lc_vel.set_linewidth(2)
        ax.add_collection(lc_vel)
        lc_vel_list.append(lc_vel)

        # Proper motion arrow scaling factor [years]
        scale_yr = 30
        dx = pm_ra.value * scale_yr   # [mas]
        dy = pm_dec.value * scale_yr  # [mas]
        start = (ra, dec)
        end = (ra + dx, dec + dy)

        # Color star marker by tangential velocity
        color_star = cmap(norm_vel(v2D.value))
        ax.plot(ra, dec, marker='*', color=color_star, markersize=12, label=f'Star {star_key}')

        # Draw proper motion vector (arrow)
        ax.annotate('', xy=end, xytext=start,
                    arrowprops=dict(arrowstyle='->', color=color_star, lw=2),
                    annotation_clip=False)

    # Add a colorbar for velocity
    cbar = plt.colorbar(lc_vel_list[-1], ax=ax)
    cbar.set_label("Projected Velocity (km/s)")
   
    # Plot the cluster center marker 
    ax.plot(ra_cm_mas, dec_cm_mas, 'k+', markersize=10, label='Central Mass')
    
    # Plot projected periapsis position [mas]
    peri_ra  = ra_cm_mas  + (-pos[0, 1] * km_to_mas)
    peri_dec = dec_cm_mas + ( pos[0, 0] * km_to_mas)
    ax.plot(peri_ra,  peri_dec, marker='d', color='black', label='Periapsis')

    # Final axis settings and display
    ax.set_title("Projected Orbit (RA/Dec, mas)")
    ax.set_xlabel("Right Ascension (mas)")
    ax.set_ylabel("Declination (mas)")
    ax.invert_xaxis()  # Sky convention: RA increases leftward
    ax.axis('equal')
    ax.grid(True)
    ax.legend()
    plt.tight_layout()

    # Save file
    companion_names = "_".join(star_keys)
    plot_name = f"two_body_orbits_{companion_names}"
    timestamp = datetime.now().strftime("%Y-%m-%d_%H-%M-%S")
    filename = rf"C:\Users\macke\OneDrive - Saint Marys University\Summer Research 2025\Plots\{plot_name}_{timestamp}.png"

    # Save the figure and show
    plt.savefig(filename, dpi=300)
    print(f"Plot saved to {filename}")
    plt.show()

stars_orbits(["A", "B"], utils, octo_utils)
```
